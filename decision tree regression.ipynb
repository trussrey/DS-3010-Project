```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

print(combined_df.info())

df = combined_df
df.head()
```

```python
# make a decision tree model to predict popularity based on these features:
# status, show_votes, type, Popularity, no. of seasons,
# genre or genre_type, production companies/countries, Air dates
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import plot_tree, DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn import utils
# Preprocessing: Drop irrelevant columns and handle missing values

df1 = df[['status_id', 'type_id', 'popularity', 'vote_average', 'number_of_seasons',
          'created_by_type_id', 'genre_type_id',
          'production_country_type_id', 'production_company_type_id',
          'language_type_id', 'spoken_language_type_id', 'network_type_id',
           'in_production', 'adult']]

# popularity vote average and number of seasons


# Split the dataset into features (X) and target variable (y)
y = (df1['number_of_seasons'] +df1['vote_average'])/2  # target variable
X = df1.drop('type_id',axis=1) # Features


label_encoder = preprocessing.LabelEncoder()
y_transformed = label_encoder.fit_transform(y)

corr = df1.corr()
plt.figure(figsize=(30, 20))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y_transformed, test_size=0.3, random_state=1000)
```

```python
# Initialize and train the Decision Tree model
clf = DecisionTreeRegressor(criterion='squared_error', random_state=1000, max_depth=10, max_leaf_nodes=80)
clf.fit(X_train, y_train)
```

```python
# Predict the response for test dataset using model
y_pred = clf.predict(X_test)
y_pred
```

```python
# Model evaluation 
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Visualize the decision tree
text_representation = tree.export_text(clf, feature_names=list(X.columns))
print(text_representation)
fig = plt.figure(figsize=(25,20))
feature_names1 = [str(i) for i in X.columns]
target_name = df['popularity']
plot_tree(clf, 
             filled=True, feature_names=feature_names1, 
             class_names=target_name)
plt.show()
```

```python
# Get feature importances
feature_importances = clf.feature_importances_
print(feature_importances)

# Get indices of features with non-zero importances
non_zero_indices = np.where(feature_importances > 0)[0]

# Get the names of features with non-zero importances
selected_features = [feature_names1[i] for i in non_zero_indices]

print("Features used in the decision tree regression model:")
print(selected_features)
print("Decision Tree Model Accuracy:", clf.score(X_test, y_test))
print("Decision Tree Model Mean Squared Error:", mse)
```
